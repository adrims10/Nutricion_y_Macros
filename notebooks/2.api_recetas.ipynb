```python
import requests
import json
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('../'))
from src import soporte_Re as sr
import dotenv 
dotenv.load_dotenv()
key = os.getenv("key")
```

```python
url = "https://tasty.p.rapidapi.com/recipes/list"

querystring = {
    "from":"0","size":"100","q":"protein"}
    
headers = {
        "x-rapidapi-key": key,
        "x-rapidapi-host": "tasty.p.rapidapi.com"
    }

response = requests.get(url, headers=headers, params=querystring)
data = response.json()

results = data['results']
```

```python
df= sr.recetas(results)
```
